```python
New_colo={
    'timestamp':'Time',
    'cnt' : 'Count',
    't1' : 'Temp_real',
    't2' : 'Temp_feel_like',
    'hum' : 'Humidity_percentage',
    'wind_speed' : 'Wind_speed_kph',
    'weather_code' :'Weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

London_data.rename(columns=New_colo,inplace=True)
```

```python
London_data
```

```python
London_data['Humidity_percentage']=London_data['Humidity_percentage']/100
```

```python
London_data.head()
```

```python
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}

weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}

London_data['season']=London_data['season'].astype('str')
London_data['Weather']=London_data['Weather'].astype('str')
```

```python
London_data.season=London_data.season.map(season_dict)
London_data.Weather=London_data.Weather.map(weather_dict)
```
